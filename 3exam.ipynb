```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'Student name': ['Vlad', 'Ilya', 'Denis', 'Alina', 'Katya'],
    'Attendance (%)': [71, 95, 85, 80, 70],
    'Mark': [65, 90, 79, 74, 64]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Attendance (%)', y='Mark', hue='Student name', s=100)
sns.regplot(data=df, x='Attendance (%)', y='Mark', scatter=False, color='red')
plt.title('corr between attendance and mark', fontsize=16)
plt.xlabel('Attendance (%)', fontsize=14)
plt.xticks(range(0, 101, 10))
plt.yticks(range(0, 101, 10))
plt.grid(True)
plt.show()
```
